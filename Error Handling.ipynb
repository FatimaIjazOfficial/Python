# ⚠️ Error Handling in Python

Error handling is used to prevent program crashes and manage unexpected situations safely.

In real-world Data Science / AI projects, errors happen in:
- Data loading
- API calls
- File handling
- Model training
- User input

Python provides `try-except` blocks to handle errors.

### 1. Basic Try-Except

Used when we expect a possible error.

```python
try:
    x = 10 / 0
    print(x)
except ZeroDivisionError:
    print("Error: Cannot divide by zero")
```

### 2. Multiple Exceptions Handling

Different types of errors can occur, so we handle them separately.

```python
try:
    num = int("abc")
    result = 10 / 0

except ValueError:
    print("Error: Invalid value conversion")

except ZeroDivisionError:
    print("Error: Division by zero")
```

### 3. Using `else` block

`else` runs only when NO error occurs.

```python
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error occurred")
else:
    print("No error occurred, result is:", x)
```

### 4. Finally Block

`finally` always runs (error or no error).
Used for cleanup operations.

```python
try:
    file = open("test.txt", "w")
    file.write("Hello")
except Exception as e:
    print("Error:", e)
finally:
    file.close()
    print("File closed successfully")
```

### 5. Catching General Exceptions

Used when we don’t know exact error type.
(Not recommended for production without reason)

```python
try:
    a = [1, 2, 3]
    print(a[10])

except Exception as e:
    print("General Error:", e)
```

### 6. Raising Custom Errors

We can manually create errors using `raise`.

```python
age = 15

if age < 18:
    raise ValueError("Age must be 18 or above")
```

### 7. Custom Exception Class

Used in real projects for validation systems.

```python
class MyCustomError(Exception):
    pass

def check_salary(salary):
    if salary < 10000:
        raise MyCustomError("Salary too low")

try:
    check_salary(5000)
except MyCustomError as e:
    print("Custom Error:", e)
```

### 8. Real-World Example (Data Input Validation)

This is how error handling is used in real Data Science projects.

```python
def divide_numbers(a, b):
    try:
        result = a / b
        return result

    except ZeroDivisionError:
        return "Cannot divide by zero"

    except TypeError:
        return "Invalid data type"

print(divide_numbers(10, 2))
print(divide_numbers(10, 0))
print(divide_numbers(10, "a"))
```

✔ Always handle specific exceptions first  
✔ Avoid empty except blocks  
✔ Use finally for cleanup  
✔ Use custom exceptions for validation  
✔ Don’t overuse try-except (fix logic first)
