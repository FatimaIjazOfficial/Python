# Object Oriented Programming (OOP)
OOP is a programming paradigm based on objects and classes.
##### - Class -> blueprint/template for creating objects.
##### - Object -> Real instance created from class.
##### - Constructor (__init__) -> Automatically runs when object is created.
##### - Instance Attributes -> Unique for each object.
##### - Class Attributes -> Shared among all objects.
##### - Dynamic Attributes -> attributes that are added to an object outside the class definition, after the object is created.
##### - Private Attributes -> variables cannot inside a class that cannot be accessesd directly from outside the class. So we use getter to access.
##### - Getter Method(@property) -> returns private attribute
##### - Setter Method(@method.setter) -> validates and sets private attribute
##### - Class Method(@classmethod) -> A method that works with the class itself using cls and is used to access or modify class attributes.
##### - Static Method(@staticmethod) -> A method that does not use self or cls and works like a normal utility function inside the class.

```python
class Student:
    university = "PU"                              # Class Attribute 

    def __init__(self, name, age, marks):
        self.name = name                          # Instance Attributes
        self.age = age
        self.marks = marks                        # Setter is used here for validation

    def display(self):
        print("Name:", self.name)                  # Instance method
        print("Age:", self.age)
        print("Marks:", self.__marks)
        print("University:", Student.university)

    @classmethod                                   # Class method
    def change_university(cls, new_uni):
        cls.university = new_uni

    @staticmethod                                  # Static method
    def info():
        print("This is Student class example")

    @property                                      # Getter method
    def marks(self):
        return self.__marks         

    @marks.setter                                  # Setter method
    def marks(self, value):
        if value < 0 or value > 100:
            print("Invalid Marks! Setting 0")
            self.__marks = 0
        else:
            self.__marks = value
```

```python
s1 = Student("Fatima", 25, 95)                       # Object 

s1.display()                                         # Instance method call

s1.city = "Islamabad"                                # Dynamic attribute
print("City:", s1.city)

print("Marks:", s1.marks)                            # Getter Call

s1.marks = 88                                        # Setter update
print("Updated Marks:", s1.marks)

Student.change_university("UET")                     # Class method call
print("University:", s1.university)

Student.info()                                       # Static method call
```

## Encapsulation

```python
class Student:

    def __init__(self, name, marks):
        self.name=name
        self.__marks=marks                # Private Attribute

    # Getter Method
    @property
    def marks(self):
        return self.__marks

    # Setter Method
    @marks.setter
    def marks(self, value):
        if value<0 or value>100:
            print("Invalid marks")
            self.__marks=0
        else:
            self.__marks=value
```

```python
s1=Student("Fatima",94)
print(f"Before Update Marks: {s1.marks}")
s1.marks=87
print(f"After Update Marks: {s1.marks}")
```

```python
s1=Student("Fatima",94)
print(f"Before Update Marks: {s1.marks}")
s1.marks=124
print(f"After Update Marks: {s1.marks}")
```

## Inheritance

#### Single Inheritance

```python
class Person:

    def __init__(self, name):
        self.name=name
        
    def show(self):
        print("Name:", self.name)

class Student(Person):

    def __init__(self, name, field):
        super().__init__(name)
        self.field=field
        
    def show_details(self):
        print(self.name, self.field)
```

```python
s1 = Student("Fatima", "Data Science")
s1.show()
s1.show_details()
```

#### Multiple Inheritance

```python
class A:
    def showA(self):
        print("A")

class B:
    def showB(self):
        print("B")

class C(A,B):
    pass    
```

```python
obj=C()
obj.showA()
obj.showB()
```

#### Multilevel Inheritance

```python
class A:
    def showA(self):
        print("A")

class B(A):
    def showB(self):
        print("B")

class C(B):
    def showC(self):
        print("C")   
```

```python
obj=C()
obj.showA()
obj.showB()
obj.showC()
```

## Polymorphism

#### Overriding
- Child class changes parent method implementation
- Same method name, different behavior in child class                            

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Bark")

class Cat(Animal):

    def sound(self):
        print("Meow")        
```

```python
animals = [Dog(), Cat()]
for animal in animals:
    animal.sound()
```

#### Overloading
- Same method name, different parameters
- NOT natively supported in Python (simulated using default args/*args)

```python
class Math:
    
    def add(self, a,b,c=0):
        return a+b+c
```

```python
m=Math()
print(m.add(2,3))
print(m.add(2,3,4))
```

## Abstraction

```python
from abc import ABC, abstractmethod

# Abstract Class
class Shape(ABC):
   
    @abstractmethod
    def area(self):
        pass

# Implementation Class
class Rectangle(Shape):

    def __init__(self,w,h):
        self.w=w
        self.h=h

    def area(self):
        return self.w * self.h
```

```python
r=Rectangle(10,5)
print(r.area())
```
