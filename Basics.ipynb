# Variables & DataTypes
A variable is a container used to store data.
Syntax:
variable_name = value

🔹 Common Data Types
| Data Type | Example |
|------------|----------|
| String     | "Fatima" |
| Integer    | 22 |
| Float      | 5.4 |
| Boolean    | True |
Python automatically assigns the correct data type.

```python
name = "Fatima"   # String
age = 22          # Int
height = 5.4      # Float
is_student = True # Boolean
```

# Type Checking & Type Conversion
The type() function is used to check the data type of a variable.
Type conversion means converting one data type into another.
Common conversions:
- str() → convert to string
- int() → convert to integer
- float() → convert to float
- bool() → convert to boolean

```python
type(age)
```

```python
str_age = str(age)
str_age
```

```python
type(str_age)
```
