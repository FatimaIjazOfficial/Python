# 📊 Data Structures in Python
Data structures are used to store and organize data efficiently.
    
| Data Structure | Ordered | Mutable | Duplicates |
|----------------|----------|----------|-------------|
| List [ ]         | Yes      | Yes      | Yes         |
| Tuple ( )        | Yes      | No       | Yes         |
| Set { }          | No       | Yes      | No          |
| Dictionary       | Yes      | Yes      | Keys: No    |

### LIST (Mutable, Ordered, Allows duplicates)

#### 🔹 Creation
- my_list = [1, 2, 3]

#### 🔹 Access
- Indexing → my_list[0]
- Negative indexing → my_list[-1]
- Slicing → my_list[1:3]

#### 🔹 Methods
- append() → add element at end
- clear() → remove all elements
- copy() → copy list
- count() → count elements
- extend() → add multiple elements
- index() → find position
- insert() → insert at position
- pop() → remove by index
- remove() → remove value
- reverse() → reverse list
- sort() → sort list

#### 🔹 Operations
- + → join lists
- * → repeat list
- in → membership check

#### 🔹 Advanced
- List comprehension → [x for x in range(10)]
- Nested lists → [[1,2],[3,4]]

```python
# creation
my_list = [10, 20, 30, 40, 50]
print("Initial list:", my_list)
```

```python
# indexing
print("First element:", my_list[0])
print("Last element:", my_list[-1])
```

```python
# index
print("Index of 30:", my_list.index(30))
```

```python
# slicing
print("Slicing [1:4]:", my_list[1:4])
print("Slicing [1:]:", my_list[1:])
print("Slicing [:4]:", my_list[:4])
```

```python
# assign
my_list[0] = 100
print("After assigning index 0 value 100:", my_list)
```

```python
# append
my_list.append(60)
print("After append 60:", my_list)
```

```python
# extend
my_list.extend([70, 80])
print("After extend [70,80]:", my_list)
```

```python
# insert
my_list.insert(1, 999)
print("After insert 999 at index 1:", my_list)
```

```python
# remove
my_list.remove(999)
print("After removing 999:", my_list)
```

```python
# pop
removed = my_list.pop()
print("Popped random element:", removed)
print("After pop:", my_list)
```

```python
# count
print("Count of 20:", my_list.count(20))
```

```python
# sort
my_list.sort()
print("After sorting:", my_list)
```

```python
# reverse
my_list.reverse()
print("After reversing:", my_list)
```

```python
# copy
copy_list = my_list.copy()
print("Copied list:", copy_list)
```

```python
# list comprehension
squares = [x*x for x in range(5)]
print("Squares using comprehension:", squares)
```

```python
# nested list
matrix = [[1,2,3],[4,5,6]]
print("Matrix:", matrix)
print("Element [0][1]:", matrix[0][1])
```

### TUPLE (Immutable, Ordered, Allows duplicates)

#### 🔹 Creation
- my_tuple = (1, 2, 3)

#### 🔹 Access
- Indexing → my_tuple[0]
- Slicing → my_tuple[1:3]

#### 🔹 Methods
- count() → count values
- index() → find position

#### 🔹 Operations
- + → combine tuples
- * → repeat tuple
- in → membership check

#### 🔹 Advanced
- Tuple unpacking → a, b = (1,2)
- Used for function returns

```python
my_tuple = (10, 20, 30, 40, 20)
print("Tuple:", my_tuple)
```

```python
# indexing
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])
```

```python
# index
print("Index of 30:", my_tuple.index(30))
```

```python
# slicing
print("Slicing [1:4]:", my_tuple[1:4])
print("Slicing [1:]:", my_tuple[1:])
print("Slicing [:4]:", my_tuple[:4])
```

```python
# count
print("Count of 20:", my_tuple.count(20))
```

```python
# operations
print("Concatenation:", my_tuple + (50, 60))
print("Repetition:", my_tuple * 2)
```

```python
# membership
print("Is 10 in tuple?", 10 in my_tuple)
```

```python
# unpacking
a, b, c, d, e = my_tuple
print("Unpacked values:", a, b, c, d, e)
```

### SET (Unordered, Mutable, No duplicates)

### 🔹 Creation
- my_set = {1, 2, 3}

### 🔹 Methods
- add() → add element
- clear() → remove all
- copy() → copy set
- difference() → A - B
- discard() → remove safely
- intersection() → common elements
- issubset() → check subset
- issuperset() → check superset
- pop() → remove random
- remove() → remove element
- union() → combine sets
- update() → add multiple elements

#### 🔹 Operations
- | → union
- & → intersection
- - → difference

#### 🔹 Advanced
- Fast membership checking → x in set
- Used for removing duplicates

```python
my_set = {1, 2, 3, 4, 4, 5}
print("Initial set (no duplicates):", my_set)
```

```python
# add
my_set.add(10)
print("After add 10:", my_set)
```

```python
# update
my_set.update([50, 30])
print("After update with list:", my_set)
my_set.update({2,6})
print("After update with set:", my_set)
my_set.update((15,43))
print("After update with tuple:", my_set)
```

```python
my_set.update("ABCD")
my_set
```

```python
my_set.update("1743")
my_set
```

```python
my_set.update([4, "A", 5.5, 2])
print("After updating list [4, 'A', 5.5, 2]:", my_set)
```

```python
# remove
my_set.remove(10)
print("After remove 10:", my_set)
```

```python
# discard
my_set.discard(100)
print("After discard 100 (no error):", my_set)
```

```python
# operations
a = {1,2,3,4}
b = {3,4,5,6}

print("Set A:", a)
print("Set B:", b)

print("Union:", a | b)
print("Intersection:", a & b)
print("Difference A-B:", a - b)
print("Symmetric difference:", a ^ b)
```

```python
# membership
print("Is 3 in set?", 3 in my_set)
```

```python
# remove duplicates example
lst = [1,1,2,2,3,4]
unique = list(set(lst))
print("Remove duplicates from list:", unique)
```

### DICTIONARY (Key:Value, Mutable, Ordered)

#### 🔹 Creation
- student = {"name": "Ali", "age": 21}

#### 🔹 Access
- student["name"]
- student.get("name")

#### 🔹 Methods
- clear() → remove all
- copy() → copy dictionary
- fromkeys() → create dict from keys
- get() → safe access
- items() → key-value pairs
- keys() → all keys
- pop() → remove key
- popitem() → remove last item
- setdefault() → set default value
- update() → update dictionary
- values() → all values

#### 🔹 Advanced
- Nested dictionary:
  {"s1": {"name": "Ali"}}
- Dictionary comprehension:
  {key: value for items in iterable}
- Iteration:
  for k, v in dict.items()

```python
student = {
    "name": "Ali",
    "age": 21,
    "course": "DS"
}
print("Initial dictionary:", student)
```

```python
# access
print("Name:", student["name"])
print("Age:", student.get("age"))
```

```python
# add/update
student["grade"] = "A"
print("After adding grade:", student)

student.update({"city": "Lahore"})
print("After update city:", student)

student.update({"like": "apple","unlike":"sugercane"})
print("After update:", student)
```

```python
# remove
removed = student.pop("age")
print("Removed age:", removed)
print("After pop:", student)
```

```python
# keys, values, items
print("Keys:", student.keys())
print("Values:", student.values())
print("Items:", student.items())
```

```python
# loop
print("Looping dictionary:")
for k, v in student.items():
    print(k, "=>", v)
```

```python
# nested dictionary
data = {
    "s1": {"name": "Ali", "marks": 80},
    "s2": {"name": "Sara", "marks": 90}
}
print("Nested dictionary:", data)
print("Student s1 name:", data["s1"]["name"])
```

```python
# dictionary comprehension
squares = {x: x*x for x in range(5)}
print("Dictionary comprehension:", squares)
```

```python
even_dict={x : x for x in range(0,10) if x%2==0 }
even_dict
```

```python
# dataset style
dataset = [
    {"name": "Ali", "age": 21},
    {"name": "Sara", "age": 22}
]
print("Dataset loop:")
for d in dataset:
    print(d)
```
