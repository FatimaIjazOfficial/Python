# 🔁 Loops in Python

Loops are used to execute a block of code multiple times.

| Loop Type | Purpose |
|------------|----------|
| for loop   | Iterates over sequences |
| while loop | Runs until condition becomes False |
| nested loop | Loop inside another loop |
| loop over list | Access list elements one by one |

## 🔹 For Loop

```python
for i in range (5):
    print(i)
```

```python
for i in range(10):
    if(i%2==0):
        print(f"{i} is an even")
    elif(i%2!=0):
        print(f"{i} is odd")
```

```python
num=10
for i in range(1,num+1):
    fact=1
    for j in range(1,i+1):
        fact=fact*j
    print(f"factorial of {i} : {fact}")
```

```python
num=20
for i in range(1,num+1):
    p=1
    if(i==1 or i==num):
        print(f"{i} is prime")
    else:   
        for j in range(2,i):
            if(i % j == 0):
                p=0
        if(p==1):
            print(f"{i} is prime")
        else:
            print(f"{i} is not prime")
```

## 🔹 While Loop

```python
count=0
while count<5:
    print(count)
    count+=1
```

## 🔹 Nested Loop

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```
