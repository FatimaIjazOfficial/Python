# 📁 File Handling in Python

File handling is used to store data permanently in files.

Python allows us to:
- Create files
- Read files
- Write files
- Append data to files
- Handle file errors safely

We use the built-in `open()` function.

# 📂 File Modes

| Mode | Meaning |
|------|--------|
| "r"  | Read (file must exist) |
| "w"  | Write (creates/overwrites file) |
| "a"  | Append (adds data at end) |
| "x"  | Create (fails if file exists) |
| "r+" | Read + Write |
| "w+" | Write + Read (overwrite) |

### Write Mode ("w")

```python
with open("file.txt", "w") as f:
    f.write("Hello Data Science\n")
    f.write("Welcome to Python File Handling\n")
```

### Append Mode ("a")

```python
with open("file.txt", "a") as f:
    f.write("This is second line (append mode)\n")
    f.write("Data Science is powerful\n")
```

### 📖 Reading Files

We can read files in multiple ways:
- read() → reads whole file
- readline() → reads one line
- readlines() → reads all lines as list

```python
with open("file.txt", "r") as f:
    content =f.read()
    print("Full File Content:\n", content)
```

```python
with open("file.txt", "r") as f:
    line1=f.readline()
    line2=f.readline()
    print("Line 1: ", line1)
    print("Line 2: ", line2)
```

```python
with open("file.txt", "r") as f:
    lines =f.readlines()
    print(lines)

print("\n")

# Loop through lines
for line in lines:
    print("Lines", line.strip())
```

### 📍 File Pointer Concept

When reading a file:
- Pointer moves line by line
- After reading, pointer is at end of file

We can reset it using `seek()`

```python
with open("file.txt", "r") as f:
    print(f.read())                # pointer at end
    f.seek(0)                      # pointer at start
    print(f.read())                # read again
```

### CHECK IF FILE EXISTS

```python
import os
if os.path.exists("file.txt"):
    print("File exists")
else:
    print("File not found")
```

### ERROR HANDLING WITH FILES

```python
try:
    with open("missing.txt", "r") as f:
        print(f.read())

except FileNotFoundError:
    print("Error: File does not exist!")

except PermissionError:
    print("Error: Permission denied!")

except Exception as e:
    print("Unknown error:", e)

finally:
    print("File operation completed")
```

## Mini Project
Student Record File System
We will:
- Save student data
- Read student data
- Append new students

```python
def add_student(name,marks):
    with open("student.txt","a") as f:
        f.write(f"{name},{marks}\n")

def show_students():
    try:
        with open("student.txt","r") as f:
            for line in f:
                name, marks=line.strip().split(",")
                print("Name:", name, "| Marks:", marks)
    except FileNotFoundError: 
        print("No student file found yet!")

#Test
add_student("Areej",80)
add_student("Fatima",90)
show_students()
```
