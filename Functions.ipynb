# Functions
A function is a reusable block of code that performs a specific task.
Instead of writing the same code again and again, we create functions and call them whenever needed.
#### Syntax

```python
def function_name(parameters):
    # code
    return value
```

## Types of Functions

#### - Built-in Functions: Functions already available in Python.

```python
print("Hello")
```

```python
list=[1,2,3]
print("Maximum number: ",max(list))
print("Length of list: ",len(list))
print("Type: ",type(list))
```

#### - User-defined Functions: Functions created by the programmer.

```python
# Function without parameters
def greetfatima():
    print(f"Hi, My name is Fatima")

greetfatima()      #call
```

```python
# Function with parameter
def greet(name):
    print(f"Hi, My name is {name}")

greet("Fatima")    #call
```

```python
# Function with parameters
def greet(name, age):
    print(f"Hi, My name is {name} and my age is {age}")

greet("Fatima", 25) #call
```

```python
# Function With Return Value
def square(num):
    return num**2

square(2)          #call
```

```python
# Function With Default Argument
def add2(a,b=2):
    return a+b

add2(5)          #call
```

```python
# Function With Type Hinting
def add(a: int, b: int) -> int:
    return a + b

print(add(5, 10)) #call
```

```python
# Function With Arbitrary Arguments(unknown)(*args)
def total(*number):
    print(sum(number))

total(1,2,3,4,5) #call
```

```python
# Function With Arbitrary Keywords Arguments(Stores keyword arguments in dictionary form.)(**args)
def info(**data):
    print(data)

info(name="Fatima", field="AI") #call
```

#### - Lambda Functions: Anonymous single-line functions.

```python
square=lambda x : x*x
square(5)
```

#### - Recursive Functions: A Function calling itself.
##### Given: factorial(1) = 1 
##### Take this part 
- factorial(n) => n * factorial(n-1) 
- factorial(5) => 5 * factorial(4)    => 5 * 24  => 120
- factorial(4) => 4 * factorial(3)    => 4 * 6   => 24 
- factorial(3) => 3 * factorial(2)    => 3 * 2   => 6 
- factorial(2) => 2 * factorial(1)    => 2 * 1   => 2

We call factorial function multiple times and then we put values at the calling place

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```
